```python
import requests
import re
import json
from bs4 import BeautifulSoup

url = "https://22f3001919.github.io/tds_project_1/"

response = requests.get(url)

match = re.search(r'const data = (\[.*?\]);', response.text,re.DOTALL)

data = json.loads(match.group(1))


links = []
for i in data:
    if i["ST_NAME"] == "ASSAM":
      links.append("https://22f3001919.github.io/tds_project_1/"+i["link"])
```

```python

extracted_data = []

pattern = r'\d{2} KALAIGAON'

for i in links:
    response = requests.get(i)
    soup = BeautifulSoup(response.content, "html.parser")
    matching = re.search(r'const data = (\[.*?\]);', response.text,re.DOTALL)
    data = json.loads(matching.group(1))
    print(data)
    for j in data:
      if re.search(pattern, j["AC"]):
        extracted_data.append(j)

#print(extracted_data)


```

```python
# 5. For the  given AC_NAME, In which year did the winning candidate have the biggest margin percentage win over the runner up? [string]
# - If there are two or more years with the same winning margin, choose the latest one

max_margin = {}

for i in max_vote_difference.keys():
    if i not in max_margin:
        max_margin[i] = 0
    
    total_votes = 0
    for j in max_vote_difference[i]:
        total_votes += j['votes']

    vote_share_first = max_vote_difference[i][0]['votes'] / total_votes
    vote_share_runner =  max_vote_difference[i][1]['votes'] / total_votes

    max_margin[i] = (vote_share_first - vote_share_runner) * 100

print(max_margin)
```

```python
# 9. For the  given AC_NAME, In how many elections did the winner get more than 50 percent of the vote? [integer]

winner_vote_share = {}

for i in max_vote_difference.keys():
    if i not in winner_vote_share:
        winner_vote_share[i] = 0
    
    total_votes = 0
    for j in max_vote_difference[i]:
        total_votes += j['votes']

    
    vote_share_first = round((max_vote_difference[i][0]['votes'] / total_votes) * 100, 2)

    if vote_share_first > 40:
        winner_vote_share[i] = vote_share_first

print(winner_vote_share)
```

```python

extract_AC_name = {}

pattern = r'^\d+\s*'

def extract_text(s):
    # Regular expression to match one or more digits at the beginning of the string, followed by a space
    pattern = r'^\d+\s*'
    # Substitute the matched pattern with an empty string
    return re.sub(pattern, '', s)

for i in links:
    response = requests.get(i)
    soup = BeautifulSoup(response.content, "html.parser")
    matching = re.search(r'const data = (\[.*?\]);', response.text,re.DOTALL)
    data = json.loads(matching.group(1))
    for j in data:
        if j["YEAR"] not in extract_AC_name:
            extract_AC_name[j["YEAR"]] = set()

        constituency_name = extract_text(j["AC"])
        extract_AC_name[j["YEAR"]].add(constituency_name)
  
print(extract_AC_name[1957])

```

```python
all_AC = set()
keys = list(extract_AC_name.keys())


for i in keys:
    all_AC = all_AC.union(extract_AC_name[i])

print(len(all_AC))
print(all_AC)

```

```python
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import time


constituency_list = []

for i in all_AC:
    constituency_list.append(i)

geolocator = Nominatim(user_agent="joemilton")

distance_between = {}

location1 = geolocator.geocode("Kalaigaon")
coords_1 = (location1.latitude, location1.longitude)

not_found = []

for i in constituency_list:
    location2 = geolocator.geocode(i)
    if location2 != None: 
        coords_2 = (location2.latitude, location2.longitude)
        distance = geodesic(coords_1, coords_2).kilometers
        if i not in distance_between:
            distance_between[i] = 0
            
        distance_between[i] = distance
    else:
        not_found.append(i)




print("final", distance_between)


```
